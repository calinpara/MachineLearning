```python
pd.pandas.set_option('display.max_columns',None) 
df = pd.read_csv('allyears2k.csv')

copy_df = df.copy()
```

# EDA

# Missing values

```python
copy_df['DepTime'] = copy_df['DepTime'].fillna(method='ffill')
copy_df['ArrTime'] = copy_df['ArrTime'].fillna(method='ffill')
copy_df['TailNum'] = copy_df['TailNum'].fillna(method='ffill')
copy_df.drop('TailNum', inplace=True, axis=1)
copy_df['ActualElapsedTime'] = copy_df['ActualElapsedTime'].fillna(method='ffill')
copy_df['ArrDelay'] = copy_df['ArrDelay'].fillna(method='ffill')
copy_df['CRSElapsedTime'] = copy_df['CRSElapsedTime'].fillna(method='ffill')
copy_df['AirTime'] = copy_df['AirTime'].fillna(method='backfill')
copy_df['DepDelay'] = copy_df['DepDelay'].fillna(method='ffill')
copy_df['Distance'] = copy_df['Distance'].fillna(method='ffill')
copy_df['TaxiIn'] = copy_df['TaxiIn'].fillna(method='backfill')
copy_df['TaxiOut'] = copy_df['TaxiOut'].fillna(method='backfill')
copy_df.drop('CancellationCode', inplace=True, axis=1)
copy_df.drop('CarrierDelay', inplace=True, axis=1)
copy_df.drop('WeatherDelay', inplace=True, axis=1)
copy_df.drop('NASDelay', inplace=True, axis=1)
copy_df.drop('SecurityDelay', inplace=True, axis=1)
copy_df.drop('LateAircraftDelay', inplace=True, axis=1)
```

# Encoding categorical variables

```python
from sklearn.preprocessing import LabelEncoder

lbl_encode = LabelEncoder()

copy_df['Dest'] = lbl_encode.fit_transform(copy_df['Dest'])
copy_df['Origin'] = lbl_encode.fit_transform(copy_df['Origin'])
copy_df['UniqueCarrier'] = lbl_encode.fit_transform(copy_df['UniqueCarrier'])
copy_df['IsArrDelayed'] = lbl_encode.fit_transform(copy_df['IsArrDelayed'])
copy_df['IsDepDelayed'] = lbl_encode.fit_transform(copy_df['IsDepDelayed'])
```

```python
analyze_report = sv.analyze([copy_df, 'DATAFRAME'],target_feat='ArrDelay')
analyze_report.show_html('2nd_analyze.html',open_browser=True)
```

# Splitting the dataset

```python
X = copy_df.drop(['ArrDelay'],axis=1)
y = copy_df['ArrDelay']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
compare_report = sv.compare([X_train,'Train'],[X_test,'Test'])
compare_report.show_html('compare.html',open_browser=True)
```

# Multiple Linear Regression

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train,y_train)

y_pred = model.predict(X_test)
```

```python
df = pd.DataFrame({'Predicted': y_pred,'Actual': y_test})
df
```

# Regression Tree

```python
from sklearn.tree import DecisionTreeRegressor

dt = DecisionTreeRegressor(min_samples_leaf=.1,random_state=1)
dt.fit(X_train, y_train)

y_pred_2 = dt.predict(X_test)
```

# Evaluation of the model

```python
from sklearn.metrics import mean_squared_error, r2_score

test_set_rmse = (np.sqrt(mean_squared_error(y_test, y_pred_2)))
test_set_r2 = r2_score(y_test, y_pred)
test_set_adj_r2 = 1-(1-r2_score(y_test, y_pred))*((len(X_test)-1)/(len(X_test)-len(X_test.columns)-1))

print('Root square mean error is: {}'.format(test_set_rmse))
print('R square is: {}'.format(test_set_r2))
print('R square adjusted is: {}'.format(test_set_adj_r2))
```
